```python
import pandas as pd
import warnings
import os
warnings.filterwarnings("ignore")
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv("YOY/Jan 2019.csv")
df_location = df[['end station name', 'end station latitude', 'end station longitude']]
df_group = df.groupby(['end station name']).count()
```

```python
df_group = df_group[['tripduration']]
df_renamed = df_group.rename(columns={'tripduration':'Trip End Count'})
df_joined_2019= pd.merge(df_location, df_renamed, on='end station name')
df_stats= df_joined_2019.drop_duplicates()
df_stats
```
