```python
import pandas as pd
import warnings
import os
warnings.filterwarnings("ignore")
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv("YOY/Jan 2019.csv")
df_location = df[['start station name', 'start station latitude', 'start station longitude']]
df_group = df.groupby(['start station name']).count()
```

```python
df_group = df_group[['tripduration']]
df_renamed = df_group.rename(columns={'tripduration':'Trip Start Count'})
df_joined_2019= pd.merge(df_location, df_renamed, on='start station name')
df_stats= df_joined_2019.drop_duplicates()
df_stats
```

```python
df_stats.to_csv("Location.csv", index=False, header=True)
```
